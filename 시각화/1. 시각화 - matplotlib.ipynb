### 틱 설정

- 틱이란 플롯이나 차트에서 축상의 위치 표시 지점을 의미하며 틱에 써진 숫자 혹은 글자를 틱 라벨이라고 한다
- 자동으로 정해지지만 수동으로 설정하려면 xticks나 yticks를 이용하면 된다.

````python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256) #np.pi = 3.14
C = np.cos(X)

plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]) #x축에 표시되는 값
plt.yticks([-1, 0, +1]) #y축에 표시되는 값
plt.show()
````

### 그리드 설정
- 틱위치를 잘보여주기 위해 그림 중간에 그리드 선을 나타내려면 grid(True)를 사용하면 된다.

````python
X = np.linspace(-np.pi, np.pi, 256) #np.pi = 3.14
C = np.cos(X)

plt.title("x축과 y축의 tick label 설정")
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi]) #x축에 표시되는 값
plt.yticks([-1, 0, +1],["Low","Zero","High"]) #y축에 표시되는 값
plt.grid(True) # 구역 표시
plt.show()
````

### 여러선 그리기
- 두가지 방법이 있는데 하나는 그냥 하나의 plot 명령어에 x,y를 순서대로 넣는 것이며 다른 하나는 n개의 plot 명령어를 사용해 n개의 그래프를 그리는 것이다.

````python
plt.title("한 개의 plot 명령어를 이용")
# 'r(컬러), s(마커), --(라인스타일)'
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--',
         [10, 20, 30, 40], [5, 10, 12, 20], 'bs-')


plt.show()
````

### 범례표시

- 여러개의 라인 플롯을 동시에 그리는 경우 선을 부가적으로 설명할 때 사용
- 자동으로 정해지지만 설정하고 싶으면 loc인수를 이용

````python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)

plt.title("legend를 표시한 플롯")

# 레이블이라는 파라미터를 무조건 넣어주어야 범례가 표시된다.
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")

# legend는 범례를 표시한다는것 loc=0은 범례의 위치를 컴퓨터에게 맡겨 정하게하는것
plt.legend(loc=0)

plt.show()
````

### x축 y축 라벨/ 타이틀
- xlabel, ylabel, title 명령어로 사용가능

````python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time") # x축의 이름 추가
plt.ylabel("amplitude") # y축의 이름 추가
plt.title("Cosine Plot")
plt.show()
````

### 연습문제


아래는 백만원을 은행에 년 3%의 복리이자로 예금했을 때 10년 동안 매년 찾을 수 있는 원리금을 계산하는 프로그램이다.

```
deposit = 1000000
year = 10
interest = 3
[deposit * (1.0 + interest / 100.0) ** i for i in range(1, year+1)]
```
이 프로그램을 이용하여 이율이 3%, 5%, 10%일 때 100만원 예금 시 10년 동안의 원리금 변화를 아래 그림과 같이 도식하는 프로그램을 작성하시오.
<img src="http://drive.google.com/uc?export=view&id=1Bc4XqY8vZA1Db9Qt5crR1J-bbAMb1xtV" style="width: 50%">

````python
deposit = 1000000
year = 10
interest = 3

y1 = [deposit * (1.0 + (interest) / 100.0) ** i for i in range(1, year+1)]
y2 = [deposit * (1.0 + (interest+2) / 100.0) ** i for i in range(1, year+1)]
y3 = [deposit * (1.0 + (interest+7) / 100.0) ** i for i in range(1, year+1)]

x = range(1, year+1)

plt.plot(x,y1,label = '이율 3%')
plt.plot(x,y2,label = '이율 5%')
plt.plot(x,y3,label = '이율 10%')

plt.xlabel('예금기간')
plt.ylabel('원리금')
plt.title('이율에 따른 원리금 차이')
plt.grid(True)
plt.legend(loc = 'best')
plt.show()
````

### bar 차트

````python
y = [2, 3, 1]
x = np.arange(len(y))
xlabel = ['가', '나', '다']
plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
````

### barh 차트

````python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))

# np.random.rand() 괄호안의 수치만큼의 난수 리스트를 출력
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
````

## 파이 차트
- 카테고리 별 값의 상대적인 비교를 해야 할 때 사용
- 하지만 추천하는 방법은 아님(대부분 바 차트로 비교)
- 원의 형태를 유지할 수 있도록 다음 명령 실행 -> plt.axis('equal')

````python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
# 색지정, 안할경우 자동 설정
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# explode는 띄워놓는 것
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90) # %%는 퍼센트로 표현하라는 뜻
plt.axis('equal')
plt.show()
````

## Axes객체와 subplot 명령
- plot 명령을 바로 사용해면 자동으로 Axes를 생성
- Figure가 행렬이고 Axes가 행렬의 원소
- 주로 2개이상의 그래프를 표현하고자 할 때 사용된다.
- 첫번째 플롯을 가리키는 숫자가 0이 아니라 1임에 주의하라.

````python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
````

`subplots` 명령으로 복수의 Axes 객체를 동시에 생성할 수도 있다. 이때는 2차원 ndarray 형태로 Axes 객체가 반환된다.

````python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")

axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")

axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")

axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout() # 그림을 조금 떨어뜨려서 표현하라
plt.show()
````
